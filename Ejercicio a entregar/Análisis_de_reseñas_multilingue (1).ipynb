```python
#modelo multilingue
# ========================================
# ANÁLISIS DE SENTIMIENTOS MULTILINGÜE
# Reseñas en Español, Inglés, Portugués, Francés, etc.
# ========================================

from transformers import pipeline
import pandas as pd
import matplotlib.pyplot as plt

# ========================================
# RESEÑAS EN MÚLTIPLES IDIOMAS
# ========================================
reseñas_multilingues = [
    # Español
    {"texto": "¡Excelente restaurante! La comida estaba deliciosa y el servicio impecable.",
     "idioma": "🇪🇸 Español"},

    {"texto": "Muy mala experiencia, la comida llegó fría y tardaron mucho.",
     "idioma": "🇪🇸 Español"},

    # Inglés
    {"texto": "Amazing food! Best restaurant I've been to in years. Highly recommended!",
     "idioma": "🇬🇧 Inglés"},

    {"texto": "Terrible service, waited over an hour for cold food. Never coming back.",
     "idioma": "🇬🇧 Inglés"},

    # Portugués
    {"texto": "Comida maravilhosa! O atendimento foi excelente e o ambiente muito agradável.",
     "idioma": "🇧🇷 Portugués"},

    {"texto": "Péssimo! A comida estava fria e o garçom foi muito grosseiro.",
     "idioma": "🇧🇷 Portugués"},

    # Francés
    {"texto": "Excellent restaurant! La cuisine était délicieuse et le service impeccable.",
     "idioma": "🇫🇷 Francés"},

    {"texto": "Très décevant, la nourriture était froide et le service lent.",
     "idioma": "🇫🇷 Francés"},

    # Italiano
    {"texto": "Fantastico! Il cibo era delizioso e il servizio eccellente.",
     "idioma": "🇮🇹 Italiano"},

    {"texto": "Pessima esperienza, il cibo era freddo e il servizio terribile.",
     "idioma": "🇮🇹 Italiano"},

    # Alemán
    {"texto": "Ausgezeichnetes Restaurant! Das Essen war köstlich und der Service perfekt.",
     "idioma": "🇩🇪 Alemán"},

    {"texto": "Sehr enttäuschend, das Essen war kalt und der Service langsam.",
     "idioma": "🇩🇪 Alemán"}
]

print("="*100)
print("🌍 ANÁLISIS DE SENTIMIENTOS MULTILINGÜE")
print("="*100)
print(f"\n📊 Total de reseñas: {len(reseñas_multilingues)}")
print(f"🗣️  Idiomas: Español, Inglés, Portugués, Francés, Italiano, Alemán\n")

# ========================================
# MODELOS MULTILINGÜES DISPONIBLES
# ========================================
print("="*100)
print("🤖 MODELOS MULTILINGÜES DISPONIBLES")
print("="*100)

modelos_info = """
┌─────────────────────────────────────────────────────────────────────────────┐
│ 1. XLM-RoBERTa (⭐ RECOMENDADO)                                             │
│    Modelo: cardiffnlp/twitter-xlm-roberta-base-sentiment                   │
│    Idiomas: 100+ idiomas                                                    │
│    Ventajas: Mejor precisión, entrenado en Twitter                         │
│    Etiquetas: Negative, Neutral, Positive                                   │
├─────────────────────────────────────────────────────────────────────────────┤
│ 2. mBERT (Multilingual BERT)                                                │
│    Modelo: nlptown/bert-base-multilingual-uncased-sentiment                │
│    Idiomas: 104 idiomas                                                     │
│    Ventajas: Da calificaciones de 1-5 estrellas                            │
│    Uso: Ideal para reviews con rating                                       │
├─────────────────────────────────────────────────────────────────────────────┤
│ 3. DistilBERT Multilingüe (MÁS RÁPIDO)                                     │
│    Modelo: lxyuan/distilbert-base-multilingual-cased-sentiments-student    │
│    Idiomas: 100+ idiomas                                                    │
│    Ventajas: 40% más rápido, usa menos memoria                             │
│    Etiquetas: negative, neutral, positive                                   │
├─────────────────────────────────────────────────────────────────────────────┤
│ 4. LaBSE (Language-agnostic BERT)                                           │
│    Modelo: Para embeddings multilingües                                     │
│    Idiomas: 109 idiomas                                                     │
│    Uso: Comparar similitud entre textos en diferentes idiomas              │
└─────────────────────────────────────────────────────────────────────────────┘
"""
print(modelos_info)

# ========================================
# OPCIÓN 1: XLM-RoBERTa (Mejor precisión)
# ========================================
print("\n" + "="*100)
print("🔄 Cargando XLM-RoBERTa (Análisis de Sentimientos Multilingüe)")
print("="*100)
print("   ⏳ Primera carga puede tardar 1-2 minutos...\n")

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-xlm-roberta-base-sentiment",
    tokenizer="cardiffnlp/twitter-xlm-roberta-base-sentiment"
)

print("✅ Modelo cargado exitosamente!\n")

# ========================================
# ANALIZAR TODAS LAS RESEÑAS
# ========================================
print("="*100)
print("🔍 ANALIZANDO RESEÑAS EN MÚLTIPLES IDIOMAS")
print("="*100)

resultados = []

for i, reseña in enumerate(reseñas_multilingues, 1):
    # Analizar sentimiento
    analisis = sentiment_analyzer(reseña['texto'])[0]

    # Mapear etiquetas
    sentimiento_map = {
        'LABEL_0': 'Negativo',
        'LABEL_1': 'Neutral',
        'LABEL_2': 'Positivo',
        'negative': 'Negativo',
        'neutral': 'Neutral',
        'positive': 'Positivo'
    }

    sentimiento = sentimiento_map.get(analisis['label'], analisis['label'])

    resultados.append({
        'ID': i,
        'Idioma': reseña['idioma'],
        'Reseña': reseña['texto'][:60] + '...' if len(reseña['texto']) > 60 else reseña['texto'],
        'Sentimiento': sentimiento,
        'Confianza (%)': round(analisis['score'] * 100, 2)
    })

    # Mostrar resultado individual
    emoji_sent = "😊" if sentimiento == "Positivo" else "😞" if sentimiento == "Negativo" else "😐"
    print(f"\n{i}. {reseña['idioma']}")
    print(f"   📝 {reseña['texto'][:70]}...")
    print(f"   {emoji_sent} {sentimiento} ({analisis['score']*100:.1f}%)")

# ========================================
# CREAR DATAFRAME Y MOSTRAR RESULTADOS
# ========================================
df = pd.DataFrame(resultados)

print("\n" + "="*100)
print("📊 TABLA COMPLETA DE RESULTADOS")
print("="*100)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 65)
print("\n" + df.to_string(index=False))

# ========================================
# ESTADÍSTICAS POR IDIOMA
# ========================================
print("\n" + "="*100)
print("📈 ESTADÍSTICAS POR IDIOMA")
print("="*100)

# Análisis por idioma
stats_por_idioma = df.groupby('Idioma').agg({
    'Sentimiento': lambda x: x.value_counts().to_dict(),
    'Confianza (%)': 'mean'
}).round(2)

print("\n🌍 Distribución de sentimientos por idioma:\n")
for idioma in df['Idioma'].unique():
    datos = df[df['Idioma'] == idioma]
    conteo = datos['Sentimiento'].value_counts()
    confianza_prom = datos['Confianza (%)'].mean()

    print(f"{idioma}:")
    for sent, cant in conteo.items():
        emoji = "😊" if sent == "Positivo" else "😞" if sent == "Negativo" else "😐"
        print(f"   {emoji} {sent}: {cant}")
    print(f"   💯 Confianza promedio: {confianza_prom:.2f}%")
    print()

# ========================================
# ESTADÍSTICAS GLOBALES
# ========================================
print("="*100)
print("📊 RESUMEN GLOBAL")
print("="*100)

conteo_global = df['Sentimiento'].value_counts()
print("\n🔢 Distribución global de sentimientos:")
for sent, cant in conteo_global.items():
    emoji = "😊" if sent == "Positivo" else "😞" if sent == "Negativo" else "😐"
    porcentaje = (cant / len(df) * 100)
    print(f"   {emoji} {sent}: {cant} ({porcentaje:.1f}%)")

confianza_global = df['Confianza (%)'].mean()
print(f"\n💯 Confianza promedio global: {confianza_global:.2f}%")

# ========================================
# VISUALIZACIÓN
# ========================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: Distribución global
colores = {'Positivo': '#2ecc71', 'Negativo': '#e74c3c', 'Neutral': '#95a5a6'}
colores_lista = [colores[sent] for sent in conteo_global.index]

ax1.bar(conteo_global.index, conteo_global.values, color=colores_lista, edgecolor='black', linewidth=1.5)
ax1.set_title('📊 Distribución Global de Sentimientos', fontsize=14, fontweight='bold', pad=15)
ax1.set_xlabel('Sentimiento', fontsize=11, fontweight='bold')
ax1.set_ylabel('Cantidad', fontsize=11, fontweight='bold')
ax1.grid(axis='y', alpha=0.3, linestyle='--')

for i, v in enumerate(conteo_global.values):
    ax1.text(i, v + 0.1, str(v), ha='center', va='bottom', fontweight='bold', fontsize=11)

# Gráfico 2: Sentimientos por idioma
idiomas_unicos = df['Idioma'].unique()
sentimientos_unicos = df['Sentimiento'].unique()

datos_idiomas = []
for idioma in idiomas_unicos:
    datos_idioma = df[df['Idioma'] == idioma]
    valores = [len(datos_idioma[datos_idioma['Sentimiento'] == sent]) for sent in sentimientos_unicos]
    datos_idiomas.append(valores)

x = range(len(idiomas_unicos))
width = 0.25

for i, sentimiento in enumerate(sentimientos_unicos):
    valores = [datos[i] for datos in datos_idiomas]
    offset = width * (i - 1)
    ax2.bar([p + offset for p in x], valores, width, label=sentimiento,
            color=colores[sentimiento], edgecolor='black', linewidth=1)

ax2.set_title('🌍 Sentimientos por Idioma', fontsize=14, fontweight='bold', pad=15)
ax2.set_xlabel('Idioma', fontsize=11, fontweight='bold')
ax2.set_ylabel('Cantidad', fontsize=11, fontweight='bold')
ax2.set_xticks(x)
ax2.set_xticklabels(idiomas_unicos, rotation=45, ha='right')
ax2.legend(title='Sentimiento', loc='upper right')
ax2.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('analisis_multilingue.png', dpi=300, bbox_inches='tight')
print("\n📁 Gráfico guardado como 'analisis_multilingue.png'")

# ========================================
# COMPARACIÓN DE MODELOS
# ========================================
print("\n" + "="*100)
print("🔬 COMPARACIÓN: ¿Cuándo usar cada modelo?")
print("="*100)

comparacion = """
┌─────────────────────────────────────────────────────────────────────────────┐
│ CASO DE USO                          │ MODELO RECOMENDADO                   │
├──────────────────────────────────────┼──────────────────────────────────────┤
│ Reseñas de redes sociales            │ ⭐ XLM-RoBERTa                       │
│ (Twitter, Instagram, Facebook)       │                                      │
├──────────────────────────────────────┼──────────────────────────────────────┤
│ Reviews con calificación (estrellas) │ mBERT (nlptown)                      │
│                                      │ → Da 1-5 estrellas directamente      │
├──────────────────────────────────────┼──────────────────────────────────────┤
│ Aplicaciones con recursos limitados  │ DistilBERT Multilingüe               │
│ (móviles, edge computing)            │ → 40% más rápido                     │
├──────────────────────────────────────┼──────────────────────────────────────┤
│ Búsqueda semántica multilingüe       │ LaBSE o sentence-transformers        │
│ (encontrar reseñas similares)        │                                      │
└────────────────────────
"""
```
