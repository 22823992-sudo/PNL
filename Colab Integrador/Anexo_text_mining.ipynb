### Ubicación Metodológica: ¿Dónde Estamos?

En el campo del Procesamiento de Lenguaje Natural existen diferentes enfoques metodológicos que han evolucionado históricamente:

**ETAPA 1: MÉTODOS BASADOS EN REGLAS**
- Fundamento: "Le decimos a la máquina exactamente qué hacer"
- Enfoque: Programación explícita con expresiones regulares y diccionarios
- Ejemplos: Buscar patrones específicos, validar formatos, extraer información estructurada

**ETAPA 2: MÉTODOS ESTADÍSTICOS/FRECUENCIALES** (AQUÍ ESTAMOS HOY)
- Fundamento: "La máquina cuenta y encuentra patrones en los números"
- Enfoque: Análisis de frecuencias y patrones estadísticos
- Herramientas: Bag of Words, TF-IDF, análisis de frecuencias
- Proyecto: Text Mining con el corpus de Hernán Casciari

**ETAPA 3: MACHINE LEARNING TRADICIONAL** (PRÓXIMO DESTINO)
- Fundamento: "La máquina aprende patrones de ejemplos"
- Enfoque: Algoritmos de clasificación y clustering
- Ejemplos: Clasificación de sentimientos, detección de spam

**ETAPAS FUTURAS**: Redes Neuronales → Modelado de Lenguaje → Transformers → Large Language Models

### Criterios de Selección Metodológica

**USAR REGLAS cuando:**
- El patrón es explícito y conocido
- Necesitamos precisión total (ej: validar emails)
- El dominio es estrecho y controlado
- Ejemplo: Extraer fechas con formato específico

**USAR ESTADÍSTICA/FRECUENCIAS cuando:**
- Queremos explorar y descubrir patrones
- El volumen de datos es grande
- Buscamos tendencias y comparaciones
- Ejemplo: ¿De qué habla más Casciari cada año?

**USAR MACHINE LEARNING cuando:**
- Necesitamos generalización a datos nuevos
- Los patrones son complejos y no obvios
- Tenemos ejemplos etiquetados para entrenar
- Ejemplo: Clasificar emails como spam/no-spam

### El Pipeline de Text Mining: Visión General

El procesamiento de textos sigue un pipeline sistemático que transforma texto no estructurado en información cuantificable:

1. **EXPLORACIÓN** → ¿Qué datos tenemos?
2. **LIMPIEZA** → ¿Qué ruido eliminar?
3. **TOKENIZACIÓN** → ¿Cómo dividir el texto?
4. **FILTRADO** → ¿Qué palabras importan?
5. **VECTORIZACIÓN** → ¿Cómo representar numéricamente?
6. **ANÁLISIS** → ¿Qué patrones emergen?
7. **VISUALIZACIÓN** → ¿Cómo comunicar hallazgos?

### Transformaciones Conceptuales Clave

**Listas → Vocabularios**
Las listas de Python se convierten en vocabularios de nuestro corpus. Cada palabra única se convierte en una dimensión de análisis.

**Diccionarios → Vectores de Frecuencia**
Los conteos de palabras se almacenan en diccionarios que posteriormente se transforman en vectores numéricos para análisis matemático.

**DataFrames → Matrices Documento-Término**
La estructura bidimensional de pandas permite representar corpus completos: cada fila es un documento, cada columna es una palabra.

**Lambda + apply() → Transformaciones Masivas**
Las funciones lambda con apply() nos permiten procesar miles de documentos aplicando operaciones de limpieza y transformación de manera eficiente.

```python
# Comparación: Función tradicional vs Lambda

# Función tradicional para normalizar frecuencias
def normalizar_frecuencia(frecuencia, total):
    return frecuencia / total

# Equivalente con lambda (más concisa)
normalizar_lambda = lambda x, total: x / total

# Ejemplo de uso en context de text mining
total_palabras_2005 = df_casciari.loc['2005'].sum()
print(f"Total de palabras analizadas en 2005: {total_palabras_2005}")

# Aplicación con función tradicional
freq_padre_normalizada = normalizar_frecuencia(df_casciari.loc['2005', 'padre'], total_palabras_2005)
print(f"Frecuencia normalizada de 'padre' (función): {freq_padre_normalizada:.3f}")

# Aplicación con lambda
freq_padre_lambda = normalizar_lambda(df_casciari.loc['2005', 'padre'], total_palabras_2005)
print(f"Frecuencia normalizada de 'padre' (lambda): {freq_padre_lambda:.3f}")
```

### Apply(): Transformación Vectorizada de Corpus

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

print("=" * 70)
print("NORMALIZACIÓN Y TRANSFORMACIÓN DE CORPUS - ANÁLISIS COMPLETO")
print("=" * 70)

# ===== RECONSTRUCCIÓN DE LA MATRIZ ORIGINAL =====
print("\n📊 MATRIZ DOCUMENTO-TÉRMINO ORIGINAL:")
corpus_casciari = {
    'Año': ['2004', '2005', '2006', '2007'],
    'padre': [45, 32, 28, 35],
    'madre': [23, 45, 18, 22],
    'fútbol': [67, 45, 52, 41],
    'escritura': [12, 28, 34, 29],
    'argentina': [31, 28, 25, 33],
    'blog': [15, 22, 18, 25]  # Añadida del ejercicio anterior
}

df_casciari = pd.DataFrame(corpus_casciari)
df_casciari = df_casciari.set_index('Año')
print(df_casciari)

# ===== NORMALIZACIÓN DE FRECUENCIAS =====
print("\n" + "=" * 50)
print("NORMALIZACIÓN POR DOCUMENTO")
print("=" * 50)

# Calcular totales por fila (documento/año)
totales_por_año = df_casciari.sum(axis=1)
print("\n📈 Total de palabras por año:")
print(totales_por_año)
print(f"\nPromedio de palabras por año: {totales_por_año.mean():.1f}")
print(f"Desviación estándar: {totales_por_año.std():.1f}")

# Aplicar normalización usando lambda y apply
print("\n🔄 Aplicando normalización con lambda y apply...")
df_normalizado = df_casciari.apply(lambda fila: fila / fila.sum(), axis=1)

print("\n📊 Matriz de frecuencias relativas (proporciones):")
print(df_normalizado.round(3))

# Verificación: cada fila debe sumar 1.0
print("\n✅ Verificación - suma de proporciones por año:")
sumas_verificacion = df_normalizado.sum(axis=1)
print(sumas_verificacion)
print(f"¿Todas las filas suman 1.0? {all(abs(suma - 1.0) < 1e-10 for suma in sumas_verificacion)}")

# ===== EJERCICIO 5: TRANSFORMACIÓN DE CORPUS =====
print("\n" + "=" * 50)
print("EJERCICIO 5: TRANSFORMACIÓN DE CORPUS")
print("=" * 50)

# Simular textos procesados (en un escenario real, estos vendrían del corpus original)
print("\n📝 Simulando textos procesados por año:")
textos_procesados = {
    '2004': "En el 2004 escribía sobre fútbol y familia con frecuencia, especialmente sobre mi padre y las tardes de domingo viendo partidos. Argentina siempre presente en mis reflexiones cotidianas.",
    '2005': "El 2005 fue el año donde mi madre comenzó a aparecer más en los textos. El fútbol seguía siendo central, pero la escritura como tema empezaba a emerger. El blog tomaba forma.",
    '2006': "Para 2006 la escritura se había vuelto más reflexiva. Menos fútbol directo, más metáfora. La figura del padre se matizaba. Argentina como concepto, no solo como lugar. El blog crecía.",
    '2007': "En 2007 todos los temas convergen: padre, madre, fútbol, escritura, Argentina, blog. Un año de síntesis donde cada palabra tiene el peso de la experiencia acumulada y la madurez narrativa."
}

# Crear DataFrame con textos
df_textos = pd.DataFrame({
    'Año': list(textos_procesados.keys()),
    'Texto_Procesado': list(textos_procesados.values())
})
df_textos = df_textos.set_index('Año')

print("Textos procesados por año:")
for año, texto in textos_procesados.items():
    print(f"\n{año}: {texto[:100]}...")

# Aplicar transformación usando lambda y apply()
print("\n🔢 Calculando longitud de textos con lambda y apply()...")

# SOLUCIÓN DEL EJERCICIO 5
df_textos['longitud_texto'] = df_textos['Texto_Procesado'].apply(lambda texto: len(texto))

print("\n📊 Resultado - Longitudes de texto por año:")
print(df_textos[['longitud_texto']])

# Análisis adicional de longitudes
print(f"\n📈 Estadísticas de longitud de texto:")
print(f"Promedio: {df_textos['longitud_texto'].mean():.1f} caracteres")
print(f"Mínimo: {df_textos['longitud_texto'].min()} caracteres ({df_textos['longitud_texto'].idxmin()})")
print(f"Máximo: {df_textos['longitud_texto'].max()} caracteres ({df_textos['longitud_texto'].idxmax()})")

# ===== ANÁLISIS AVANZADO CON TRANSFORMACIONES =====
print("\n" + "=" * 50)
print("ANÁLISIS AVANZADO CON TRANSFORMACIONES")
print("=" * 50)

# Combinar frecuencias normalizadas con longitudes
df_completo = df_normalizado.copy()
df_completo['longitud_texto'] = df_textos['longitud_texto']
df_completo['total_palabras'] = totales_por_año

print("\n🔍 Matriz completa (frecuencias normalizadas + métricas de texto):")
print(df_completo.round(3))

# Análisis de correlación entre longitud y frecuencias
print("\n📊 Correlación entre longitud de texto y uso de palabras:")
correlaciones_longitud = df_completo.corr()['longitud_texto'].sort_values(key=abs, ascending=False)
print(correlaciones_longitud.round(3))

# Identificar patrones
palabra_mas_correlacionada = correlaciones_longitud.drop(['longitud_texto', 'total_palabras']).abs().idxmax()
correlacion_max = correlaciones_longitud[palabra_mas_correlacionada]

print(f"\n🎯 La palabra más correlacionada con longitud de texto es:")
print(f"   '{palabra_mas_correlacionada}' (correlación: {correlacion_max:.3f})")

# ===== TRANSFORMACIONES ADICIONALES CON LAMBDA =====
print("\n" + "=" * 50)
print("TRANSFORMACIONES ADICIONALES CON LAMBDA")
print("=" * 50)

# 1. Densidad léxica (palabras clave / total palabras)
print("\n1️⃣ Calculando densidad léxica con lambda:")
df_completo['densidad_lexica'] = df_completo.apply(
    lambda fila: fila['total_palabras'] / fila['longitud_texto'] * 100
    if fila['longitud_texto'] > 0 else 0, axis=1
)

print("Densidad léxica (palabras por 100 caracteres):")
for año in df_completo.index:
    densidad = df_completo.loc[año, 'densidad_lexica']
    print(f"  {año}: {densidad:.2f}")

# 2. Score de diversidad temática (usando entropía)
print("\n2️⃣ Calculando diversidad temática con lambda:")
def shannon_entropy(proporciones):
    """Calcular entropía de Shannon para medir diversidad"""
    proporciones = proporciones[proporciones > 0]  # Evitar log(0)
    return -sum(p * np.log2(p) for p in proporciones)

# Aplicar solo a las columnas de palabras (excluyendo métricas)
palabras_cols = ['padre', 'madre', 'fútbol', 'escritura', 'argentina', 'blog']
df_completo['diversidad_tematica'] = df_normalizado[palabras_cols].apply(
    lambda fila: shannon_entropy(fila), axis=1
)

print("Diversidad temática (entropía de Shannon):")
for año in df_completo.index:
    diversidad = df_completo.loc[año, 'diversidad_tematica']
    print(f"  {año}: {diversidad:.3f} bits")

# 3. Transformación de escalamiento (normalización min-max)
print("\n3️⃣ Aplicando escalamiento min-max con lambda:")
df_escalado = df_casciari.apply(
    lambda col: (col - col.min()) / (col.max() - col.min())
    if (col.max() - col.min()) > 0 else col, axis=0
)

print("Valores escalados (0-1) por palabra:")
print(df_escalado.round(3))

# ===== VISUALIZACIÓN AVANZADA =====
print("\n📈 Generando visualizaciones avanzadas...")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Análisis de Transformaciones del Corpus Casciari', fontsize=16, fontweight='bold')

# 1. Comparación frecuencias absolutas vs relativas
ax1 = axes[0, 0]
df_casciari['fútbol'].plot(kind='bar', ax=ax1, color='lightblue', alpha=0.7, label='Absoluto')
ax1_twin = ax1.twinx()
df_normalizado['fútbol'].plot(kind='bar', ax=ax1_twin, color='red', alpha=0.7, label='Relativo')
ax1.set_title('Fútbol: Frecuencias Absolutas vs Relativas')
ax1.set_ylabel('Frecuencia Absoluta', color='blue')
ax1_twin.set_ylabel('Frecuencia Relativa', color='red')

# 2. Evolución de longitud de texto
ax2 = axes[0, 1]
df_completo['longitud_texto'].plot(kind='line', marker='o', ax=ax2, color='green', linewidth=3)
ax2.set_title('Evolución de Longitud de Texto')
ax2.set_ylabel('Caracteres')
ax2.grid(True, alpha=0.3)

# 3. Diversidad temática por año
ax3 = axes[0, 2]
df_completo['diversidad_tematica'].plot(kind='bar', ax=ax3, color='purple', alpha=0.7)
ax3.set_title('Diversidad Temática por Año')
ax3.set_ylabel('Entropía (bits)')

# 4. Heatmap de frecuencias normalizadas
ax4 = axes[1, 0]
sns.heatmap(df_normalizado.T, annot=True, fmt='.2f', cmap='YlOrRd', ax=ax4)
ax4.set_title('Heatmap Frecuencias Normalizadas')

# 5. Correlación longitud-palabras
ax5 = axes[1, 1]
correlaciones_plot = correlaciones_longitud.drop(['longitud_texto', 'total_palabras'])
correlaciones_plot.plot(kind='barh', ax=ax5, color=['red' if x < 0 else 'green' for x in correlaciones_plot])
ax5.set_title('Correlación Longitud-Palabras')
ax5.axvline(x=0, color='black', linestyle='--', alpha=0.5)

# 6. Densidad léxica vs diversidad temática
ax6 = axes[1, 2]
scatter = ax6.scatter(df_completo['densidad_lexica'], df_completo['diversidad_tematica'],
                     c=range(len(df_completo)), cmap='viridis', s=100, alpha=0.7)
for i, año in enumerate(df_completo.index):
    ax6.annotate(año, (df_completo.loc[año, 'densidad_lexica'],
                      df_completo.loc[año, 'diversidad_tematica']),
                xytext=(5, 5), textcoords='offset points')
ax6.set_xlabel('Densidad Léxica')
ax6.set_ylabel('Diversidad Temática')
ax6.set_title('Densidad vs Diversidad')

plt.tight_layout()
plt.show()

# ===== RESUMEN DE TRANSFORMACIONES =====
print("\n" + "=" * 70)
print("RESUMEN DE TRANSFORMACIONES APLICADAS")
print("=" * 70)

transformaciones_resumen = {
    'Normalización por documento': 'lambda fila: fila / fila.sum()',
    'Longitud de texto': 'lambda texto: len(texto)',
    'Densidad léxica': 'lambda fila: fila["total_palabras"] / fila["longitud_texto"] * 100',
    'Diversidad temática': 'lambda fila: shannon_entropy(fila)',
    'Escalamiento min-max': 'lambda col: (col - col.min()) / (col.max() - col.min())'
}

print("\n🔧 Transformaciones implementadas:")
for nombre, formula in transformaciones_resumen.items():
    print(f"  • {nombre}: {formula}")

# Tabla final con todas las métricas
print(f"\n📊 TABLA FINAL - TODAS LAS MÉTRICAS:")
columnas_finales = ['total_palabras', 'longitud_texto', 'densidad_lexica', 'diversidad_tematica']
tabla_final = df_completo[columnas_finales].round(3)
print(tabla_final)

print(f"\n✅ Análisis de transformaciones completado exitosamente!")
print(f"📈 Se aplicaron {len(transformaciones_resumen)} transformaciones diferentes usando lambda y apply()")
print("=" * 70)
```

Respuestas:
1- Se recomienda usar el enfoque basado en reglas cuando se trabaja con vocabulario altamentevespecializado, cuando el caudal de datos es relativamente pequeño y bien estructurados, casos en donde la precisión es más importante que la cobertura. Paralelamente, se recomienda utilizar el enfoque estadístico frecuencial cuando se trabaja con caudales masivos de datos, cuando se trabaja con alta variabilidad lingüistica, con aplicaciones que requieren ranking y scoring ya que se trata de un enfoque cuantitativo, cuando se busca detectar patrones no evidentes u obvios. En la practica moderna se intenta utilizar modelos híbridos que combinen ambos enfoques.

El enfoque estadístico-frecuencial se basa en el análisis cuantitativo de patrones linguísticos a través de la observación empírica de datos textuales. Utiliza medidas estadísticas derivadas de la frecuencia de aparición de palabras, secuencias y patrones para inferir significado y estructura. Este enfoque se basa en que las palabras siguen distribuciones predecibles, la frecuencia relativa indica importancia semántica y que los patrones emergen de los datos, no de reglas predefinidas.

2- El Text Mining abarca tanto el proceso secuencial que conforma al Pipeline como los Insights que se pueden inferir y obtener a partir del proceso y las metodologías utilizadas. El pipeline parte de la obtención del texto crudo o raw text. El primer paso es la organización y almacenamiento de la información a trabajar para alcanzar un acceso eficiente a la misma. El segundo paso es el preprocesado limpieza y normalización. Abarca desde eliminar signos de puntuación como mayúsculas y stopwords, hasta la normalización de caracteres'áéíóú' y el reemplazo de'x': 'por', '@': 'arroba', +: más, etc. El objetivo es la standarización de los textos a manipular y la disminución de ruido para avanzar en el análisis. El siguiente paso es la tokenización y segmentación, donde se determinan la cantidad de tokens por documento y se calcula la frecuencia de cada uno; el propósito es la conversión del  texto en unidades atómicas procesables. El siguiente paso es la VECTORIZACIÓN o REPRESENTACIÓN NUMÉRICA. El propósito es la  representación matemática para llevar a cabo su análisis cuantitativo. Luego se realiza el ANÁLISIS ESTADÍSTICO cuyo propósito es la extracción de patrones y relaciones complejas.


3- Se espera obtener como insights la evolución estlística y narrativa, evolución léxica; palabras más utilizadas, palabras menos utilizadas o lo que es lo mismo, cuantificación léxica. Otro insight es el análisis temático evolutivo, núcleos semánticos. Una geografía emocional de las palabras que describen aquellas emociones que prevalecen mayormente.

Es importante el enfoque frecuencial a la hora de identificar patrones, obtener insights que luego nos permitan ir profundizando en el análisis más profundo y detallado a través de modelos híbridos que complementen un enfoque con el otro.
