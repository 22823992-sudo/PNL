### Ubicaci√≥n Metodol√≥gica: ¬øD√≥nde Estamos?

En el campo del Procesamiento de Lenguaje Natural existen diferentes enfoques metodol√≥gicos que han evolucionado hist√≥ricamente:

**ETAPA 1: M√âTODOS BASADOS EN REGLAS**
- Fundamento: "Le decimos a la m√°quina exactamente qu√© hacer"
- Enfoque: Programaci√≥n expl√≠cita con expresiones regulares y diccionarios
- Ejemplos: Buscar patrones espec√≠ficos, validar formatos, extraer informaci√≥n estructurada

**ETAPA 2: M√âTODOS ESTAD√çSTICOS/FRECUENCIALES** (AQU√ç ESTAMOS HOY)
- Fundamento: "La m√°quina cuenta y encuentra patrones en los n√∫meros"
- Enfoque: An√°lisis de frecuencias y patrones estad√≠sticos
- Herramientas: Bag of Words, TF-IDF, an√°lisis de frecuencias
- Proyecto: Text Mining con el corpus de Hern√°n Casciari

**ETAPA 3: MACHINE LEARNING TRADICIONAL** (PR√ìXIMO DESTINO)
- Fundamento: "La m√°quina aprende patrones de ejemplos"
- Enfoque: Algoritmos de clasificaci√≥n y clustering
- Ejemplos: Clasificaci√≥n de sentimientos, detecci√≥n de spam

**ETAPAS FUTURAS**: Redes Neuronales ‚Üí Modelado de Lenguaje ‚Üí Transformers ‚Üí Large Language Models

### Criterios de Selecci√≥n Metodol√≥gica

**USAR REGLAS cuando:**
- El patr√≥n es expl√≠cito y conocido
- Necesitamos precisi√≥n total (ej: validar emails)
- El dominio es estrecho y controlado
- Ejemplo: Extraer fechas con formato espec√≠fico

**USAR ESTAD√çSTICA/FRECUENCIAS cuando:**
- Queremos explorar y descubrir patrones
- El volumen de datos es grande
- Buscamos tendencias y comparaciones
- Ejemplo: ¬øDe qu√© habla m√°s Casciari cada a√±o?

**USAR MACHINE LEARNING cuando:**
- Necesitamos generalizaci√≥n a datos nuevos
- Los patrones son complejos y no obvios
- Tenemos ejemplos etiquetados para entrenar
- Ejemplo: Clasificar emails como spam/no-spam

## De Python a NLP: Conectando Conceptos Fundamentales

### El Pipeline de Text Mining: Visi√≥n General

El procesamiento de textos sigue un pipeline sistem√°tico que transforma texto no estructurado en informaci√≥n cuantificable:

1. **EXPLORACI√ìN** ‚Üí ¬øQu√© datos tenemos?
2. **LIMPIEZA** ‚Üí ¬øQu√© ruido eliminar?
3. **TOKENIZACI√ìN** ‚Üí ¬øC√≥mo dividir el texto?
4. **FILTRADO** ‚Üí ¬øQu√© palabras importan?
5. **VECTORIZACI√ìN** ‚Üí ¬øC√≥mo representar num√©ricamente?
6. **AN√ÅLISIS** ‚Üí ¬øQu√© patrones emergen?
7. **VISUALIZACI√ìN** ‚Üí ¬øC√≥mo comunicar hallazgos?

### Transformaciones Conceptuales Clave

**Listas ‚Üí Vocabularios**
Las listas de Python se convierten en vocabularios de nuestro corpus. Cada palabra √∫nica se convierte en una dimensi√≥n de an√°lisis.

**Diccionarios ‚Üí Vectores de Frecuencia**
Los conteos de palabras se almacenan en diccionarios que posteriormente se transforman en vectores num√©ricos para an√°lisis matem√°tico.

**DataFrames ‚Üí Matrices Documento-T√©rmino**
La estructura bidimensional de pandas permite representar corpus completos: cada fila es un documento, cada columna es una palabra.

**Lambda + apply() ‚Üí Transformaciones Masivas**
Las funciones lambda con apply() nos permiten procesar miles de documentos aplicando operaciones de limpieza y transformaci√≥n de manera eficiente.

### Diccionarios: Vectores de Frecuencia

Los diccionarios representan la estructura fundamental para contar frecuencias de palabras. En text mining, cada diccionario puede representar un documento como vector de frecuencias t√©rmino-valor.

```python
# Ejemplo: Representando un documento como vector de frecuencias
documento_casciari = {
    "t√≠tulo": "M√°s respeto, que soy tu madre",
    "a√±o_publicacion": 2005,
    "temas_principales": ["familia", "humor", "argentina"],
    "palabras_frecuentes": {
        "madre": 45,
        "hijo": 32,
        "casa": 28,
        "familia": 23,
        "argentina": 15
    }
}

# Acceso a informaci√≥n estructurada
print("An√°lisis del documento:", documento_casciari["t√≠tulo"])
print("A√±o:", documento_casciari["a√±o_publicacion"])
print("Palabra m√°s frecuente:", max(documento_casciari["palabras_frecuentes"],
                                   key=documento_casciari["palabras_frecuentes"].get))

# Modificaci√≥n din√°mica (simulando actualizaci√≥n de conteos)
documento_casciari["palabras_frecuentes"]["humor"] = 41
print("Frecuencia de 'humor':", documento_casciari["palabras_frecuentes"]["humor"])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

print("=" * 70)
print("NORMALIZACI√ìN Y TRANSFORMACI√ìN DE CORPUS - AN√ÅLISIS COMPLETO")
print("=" * 70)

# ===== RECONSTRUCCI√ìN DE LA MATRIZ ORIGINAL =====
print("\nüìä MATRIZ DOCUMENTO-T√âRMINO ORIGINAL:")
corpus_casciari = {
    'A√±o': ['2004', '2005', '2006', '2007'],
    'padre': [45, 32, 28, 35],
    'madre': [23, 45, 18, 22],
    'f√∫tbol': [67, 45, 52, 41],
    'escritura': [12, 28, 34, 29],
    'argentina': [31, 28, 25, 33],
    'blog': [15, 22, 18, 25]  # A√±adida del ejercicio anterior
}

df_casciari = pd.DataFrame(corpus_casciari)
df_casciari = df_casciari.set_index('A√±o')
print(df_casciari)

# ===== NORMALIZACI√ìN DE FRECUENCIAS =====
print("\n" + "=" * 50)
print("NORMALIZACI√ìN POR DOCUMENTO")
print("=" * 50)

# Calcular totales por fila (documento/a√±o)
totales_por_a√±o = df_casciari.sum(axis=1)
print("\nüìà Total de palabras por a√±o:")
print(totales_por_a√±o)
print(f"\nPromedio de palabras por a√±o: {totales_por_a√±o.mean():.1f}")
print(f"Desviaci√≥n est√°ndar: {totales_por_a√±o.std():.1f}")

# Aplicar normalizaci√≥n usando lambda y apply
print("\nüîÑ Aplicando normalizaci√≥n con lambda y apply...")
df_normalizado = df_casciari.apply(lambda fila: fila / fila.sum(), axis=1)

print("\nüìä Matriz de frecuencias relativas (proporciones):")
print(df_normalizado.round(3))

# Verificaci√≥n: cada fila debe sumar 1.0
print("\n‚úÖ Verificaci√≥n - suma de proporciones por a√±o:")
sumas_verificacion = df_normalizado.sum(axis=1)
print(sumas_verificacion)
print(f"¬øTodas las filas suman 1.0? {all(abs(suma - 1.0) < 1e-10 for suma in sumas_verificacion)}")

# ===== EJERCICIO 5: TRANSFORMACI√ìN DE CORPUS =====
print("\n" + "=" * 50)
print("EJERCICIO 5: TRANSFORMACI√ìN DE CORPUS")
print("=" * 50)

# Simular textos procesados (en un escenario real, estos vendr√≠an del corpus original)
print("\nüìù Simulando textos procesados por a√±o:")
textos_procesados = {
    '2004': "En el 2004 escrib√≠a sobre f√∫tbol y familia con frecuencia, especialmente sobre mi padre y las tardes de domingo viendo partidos. Argentina siempre presente en mis reflexiones cotidianas.",
    '2005': "El 2005 fue el a√±o donde mi madre comenz√≥ a aparecer m√°s en los textos. El f√∫tbol segu√≠a siendo central, pero la escritura como tema empezaba a emerger. El blog tomaba forma.",
    '2006': "Para 2006 la escritura se hab√≠a vuelto m√°s reflexiva. Menos f√∫tbol directo, m√°s met√°fora. La figura del padre se matizaba. Argentina como concepto, no solo como lugar. El blog crec√≠a.",
    '2007': "En 2007 todos los temas convergen: padre, madre, f√∫tbol, escritura, Argentina, blog. Un a√±o de s√≠ntesis donde cada palabra tiene el peso de la experiencia acumulada y la madurez narrativa."
}

# Crear DataFrame con textos
df_textos = pd.DataFrame({
    'A√±o': list(textos_procesados.keys()),
    'Texto_Procesado': list(textos_procesados.values())
})
df_textos = df_textos.set_index('A√±o')

print("Textos procesados por a√±o:")
for a√±o, texto in textos_procesados.items():
    print(f"\n{a√±o}: {texto[:100]}...")

# Aplicar transformaci√≥n usando lambda y apply()
print("\nüî¢ Calculando longitud de textos con lambda y apply()...")

# SOLUCI√ìN DEL EJERCICIO 5
df_textos['longitud_texto'] = df_textos['Texto_Procesado'].apply(lambda texto: len(texto))

print("\nüìä Resultado - Longitudes de texto por a√±o:")
print(df_textos[['longitud_texto']])

# An√°lisis adicional de longitudes
print(f"\nüìà Estad√≠sticas de longitud de texto:")
print(f"Promedio: {df_textos['longitud_texto'].mean():.1f} caracteres")
print(f"M√≠nimo: {df_textos['longitud_texto'].min()} caracteres ({df_textos['longitud_texto'].idxmin()})")
print(f"M√°ximo: {df_textos['longitud_texto'].max()} caracteres ({df_textos['longitud_texto'].idxmax()})")

# ===== AN√ÅLISIS AVANZADO CON TRANSFORMACIONES =====
print("\n" + "=" * 50)
print("AN√ÅLISIS AVANZADO CON TRANSFORMACIONES")
print("=" * 50)

# Combinar frecuencias normalizadas con longitudes
df_completo = df_normalizado.copy()
df_completo['longitud_texto'] = df_textos['longitud_texto']
df_completo['total_palabras'] = totales_por_a√±o

print("\nüîç Matriz completa (frecuencias normalizadas + m√©tricas de texto):")
print(df_completo.round(3))

# An√°lisis de correlaci√≥n entre longitud y frecuencias
print("\nüìä Correlaci√≥n entre longitud de texto y uso de palabras:")
correlaciones_longitud = df_completo.corr()['longitud_texto'].sort_values(key=abs, ascending=False)
print(correlaciones_longitud.round(3))

# Identificar patrones
palabra_mas_correlacionada = correlaciones_longitud.drop(['longitud_texto', 'total_palabras']).abs().idxmax()
correlacion_max = correlaciones_longitud[palabra_mas_correlacionada]

print(f"\nüéØ La palabra m√°s correlacionada con longitud de texto es:")
print(f"   '{palabra_mas_correlacionada}' (correlaci√≥n: {correlacion_max:.3f})")

# ===== TRANSFORMACIONES ADICIONALES CON LAMBDA =====
print("\n" + "=" * 50)
print("TRANSFORMACIONES ADICIONALES CON LAMBDA")
print("=" * 50)

# 1. Densidad l√©xica (palabras clave / total palabras)
print("\n1Ô∏è‚É£ Calculando densidad l√©xica con lambda:")
df_completo['densidad_lexica'] = df_completo.apply(
    lambda fila: fila['total_palabras'] / fila['longitud_texto'] * 100
    if fila['longitud_texto'] > 0 else 0, axis=1
)

print("Densidad l√©xica (palabras por 100 caracteres):")
for a√±o in df_completo.index:
    densidad = df_completo.loc[a√±o, 'densidad_lexica']
    print(f"  {a√±o}: {densidad:.2f}")

# 2. Score de diversidad tem√°tica (usando entrop√≠a)
print("\n2Ô∏è‚É£ Calculando diversidad tem√°tica con lambda:")
def shannon_entropy(proporciones):
    """Calcular entrop√≠a de Shannon para medir diversidad"""
    proporciones = proporciones[proporciones > 0]  # Evitar log(0)
    return -sum(p * np.log2(p) for p in proporciones)

# Aplicar solo a las columnas de palabras (excluyendo m√©tricas)
palabras_cols = ['padre', 'madre', 'f√∫tbol', 'escritura', 'argentina', 'blog']
df_completo['diversidad_tematica'] = df_normalizado[palabras_cols].apply(
    lambda fila: shannon_entropy(fila), axis=1
)

print("Diversidad tem√°tica (entrop√≠a de Shannon):")
for a√±o in df_completo.index:
    diversidad = df_completo.loc[a√±o, 'diversidad_tematica']
    print(f"  {a√±o}: {diversidad:.3f} bits")

# 3. Transformaci√≥n de escalamiento (normalizaci√≥n min-max)
print("\n3Ô∏è‚É£ Aplicando escalamiento min-max con lambda:")
df_escalado = df_casciari.apply(
    lambda col: (col - col.min()) / (col.max() - col.min())
    if (col.max() - col.min()) > 0 else col, axis=0
)

print("Valores escalados (0-1) por palabra:")
print(df_escalado.round(3))

# ===== VISUALIZACI√ìN AVANZADA =====
print("\nüìà Generando visualizaciones avanzadas...")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('An√°lisis de Transformaciones del Corpus Casciari', fontsize=16, fontweight='bold')

# 1. Comparaci√≥n frecuencias absolutas vs relativas
ax1 = axes[0, 0]
df_casciari['f√∫tbol'].plot(kind='bar', ax=ax1, color='lightblue', alpha=0.7, label='Absoluto')
ax1_twin = ax1.twinx()
df_normalizado['f√∫tbol'].plot(kind='bar', ax=ax1_twin, color='red', alpha=0.7, label='Relativo')
ax1.set_title('F√∫tbol: Frecuencias Absolutas vs Relativas')
ax1.set_ylabel('Frecuencia Absoluta', color='blue')
ax1_twin.set_ylabel('Frecuencia Relativa', color='red')

# 2. Evoluci√≥n de longitud de texto
ax2 = axes[0, 1]
df_completo['longitud_texto'].plot(kind='line', marker='o', ax=ax2, color='green', linewidth=3)
ax2.set_title('Evoluci√≥n de Longitud de Texto')
ax2.set_ylabel('Caracteres')
ax2.grid(True, alpha=0.3)

# 3. Diversidad tem√°tica por a√±o
ax3 = axes[0, 2]
df_completo['diversidad_tematica'].plot(kind='bar', ax=ax3, color='purple', alpha=0.7)
ax3.set_title('Diversidad Tem√°tica por A√±o')
ax3.set_ylabel('Entrop√≠a (bits)')

# 4. Heatmap de frecuencias normalizadas
ax4 = axes[1, 0]
sns.heatmap(df_normalizado.T, annot=True, fmt='.2f', cmap='YlOrRd', ax=ax4)
ax4.set_title('Heatmap Frecuencias Normalizadas')

# 5. Correlaci√≥n longitud-palabras
ax5 = axes[1, 1]
correlaciones_plot = correlaciones_longitud.drop(['longitud_texto', 'total_palabras'])
correlaciones_plot.plot(kind='barh', ax=ax5, color=['red' if x < 0 else 'green' for x in correlaciones_plot])
ax5.set_title('Correlaci√≥n Longitud-Palabras')
ax5.axvline(x=0, color='black', linestyle='--', alpha=0.5)

# 6. Densidad l√©xica vs diversidad tem√°tica
ax6 = axes[1, 2]
scatter = ax6.scatter(df_completo['densidad_lexica'], df_completo['diversidad_tematica'],
                     c=range(len(df_completo)), cmap='viridis', s=100, alpha=0.7)
for i, a√±o in enumerate(df_completo.index):
    ax6.annotate(a√±o, (df_completo.loc[a√±o, 'densidad_lexica'],
                      df_completo.loc[a√±o, 'diversidad_tematica']),
                xytext=(5, 5), textcoords='offset points')
ax6.set_xlabel('Densidad L√©xica')
ax6.set_ylabel('Diversidad Tem√°tica')
ax6.set_title('Densidad vs Diversidad')

plt.tight_layout()
plt.show()

# ===== RESUMEN DE TRANSFORMACIONES =====
print("\n" + "=" * 70)
print("RESUMEN DE TRANSFORMACIONES APLICADAS")
print("=" * 70)

transformaciones_resumen = {
    'Normalizaci√≥n por documento': 'lambda fila: fila / fila.sum()',
    'Longitud de texto': 'lambda texto: len(texto)',
    'Densidad l√©xica': 'lambda fila: fila["total_palabras"] / fila["longitud_texto"] * 100',
    'Diversidad tem√°tica': 'lambda fila: shannon_entropy(fila)',
    'Escalamiento min-max': 'lambda col: (col - col.min()) / (col.max() - col.min())'
}

print("\nüîß Transformaciones implementadas:")
for nombre, formula in transformaciones_resumen.items():
    print(f"  ‚Ä¢ {nombre}: {formula}")

# Tabla final con todas las m√©tricas
print(f"\nüìä TABLA FINAL - TODAS LAS M√âTRICAS:")
columnas_finales = ['total_palabras', 'longitud_texto', 'densidad_lexica', 'diversidad_tematica']
tabla_final = df_completo[columnas_finales].round(3)
print(tabla_final)

print(f"\n‚úÖ An√°lisis de transformaciones completado exitosamente!")
print(f"üìà Se aplicaron {len(transformaciones_resumen)} transformaciones diferentes usando lambda y apply()")
print("=" * 70)
```

El enfoque estad√≠stico-frecuencial se basa en el an√°lisis cuantitativo de patrones lingu√≠sticos a trav√©s de la observaci√≥n emp√≠rica de datos textuales. Utiliza medidas estad√≠sticas derivadas de la frecuencia de aparici√≥n de palabras, secuencias y patrones para inferir significado y estructura. El pipeline parte de la obtenci√≥n del texto crudo o raw text. El primer paso es la organizaci√≥n y almacenamiento de la informaci√≥n a trabajar para alcanzar un acceso eficiente a la misma. El segundo paso es el preprocesado limpieza y normalizaci√≥n. Abarca desde eliminar signos de puntuaci√≥n como may√∫sculas y stopwords, hasta la normalizaci√≥n de caracteres'√°√©√≠√≥√∫' y el reemplazo de'x': 'por', '@': 'arroba', +: m√°s, etc. El objetivo es la standarizaci√≥n de los textos a manipular y la disminuci√≥n de ruido para avanzar en el an√°lisis. El siguiente paso es la tokenizaci√≥n y segmentaci√≥n, donde se determinan la cantidad de tokens por documento y se calcula la frecuencia de cada uno; el prop√≥sito es la conversi√≥n del  texto en unidades at√≥micas procesables. El siguiente paso es la VECTORIZACI√ìN o REPRESENTACI√ìN NUM√âRICA. El prop√≥sito es la  representaci√≥n matem√°tica para llevar a cabo su an√°lisis cuantitativo. Luego se realiza el AN√ÅLISIS ESTAD√çSTICO cuyo prop√≥sito es la extracci√≥n de patrones y relaciones complejas.

Otro insight es el an√°lisis tem√°tico evolutivo, n√∫cleos sem√°nticos. Una geograf√≠a emocional de las palabras que describen aquellas emociones que prevalecen mayormente.

Es importante el enfoque frecuencial a la hora de identificar patrones, obtener insights que luego nos permitan ir profundizando en el an√°lisis m√°s profundo y detallado a trav√©s de modelos h√≠bridos que complementen un enfoque con el otro.
