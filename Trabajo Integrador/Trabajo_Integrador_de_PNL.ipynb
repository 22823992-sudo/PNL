```python
import pandas as pd
import numpy as np
import pickle
import re
import string
import time
import os
from collections import Counter
from datetime import datetime
import unicodedata
import random
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')


path = "/content/drive/MyDrive/Programaci√≥n de lenguaje/Trabajo Integrador PNL/mi_corpus/raw_texts"
if os.path.exists(path):
    files = [f for f in os.listdir(path) if f.endswith('.txt')]
    print(f"‚úÖ {len(files)} Archivos encontrados")
    print(files[:5])  # Mostrar los 5 primeros
else:
    print("‚ùå No se encontraron archivos")

class CorpusBuilder:

    def __init__(self, base_dir="/content/drive/MyDrive/Programaci√≥n de lenguaje/Trabajo Integrador PNL/mi_corpus"):
        self.base_dir = base_dir
        self.raw_texts_dir = os.path.join(base_dir, "raw_texts")
        self.metadata_file = os.path.join(base_dir, "metadata.csv")
        self.setup_directories()

    def setup_directories(self):
        """Crea la estructura de directorios necesaria"""
        os.makedirs(self.raw_texts_dir, exist_ok=True)
        print(f"‚úÖ Estructura verificada en: {self.base_dir}")

    def clean_filename(self, text):
        """Limpia texto para nombre de archivo"""
        text = unicodedata.normalize('NFKD', text)
        text = ''.join([c for c in text if not unicodedata.combining(c)])
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
        text = re.sub(r'\s+', '_', text.strip())
        return text.lower()

    def count_words(self, text):
        """Cuenta palabras aproximadas en el texto"""
        words = re.findall(r'\b\w+\b', text.lower())
        return len(words)

    def scan_existing_corpus(self):
        """Escanea el corpus existente y crea metadata basado en archivos encontrados"""
        print("\nüîç Escaneando corpus existente...")

        if not os.path.exists(self.raw_texts_dir):
            print("‚ùå Directorio raw_texts no encontrado")
            return False

        # Buscar archivos .txt en el directorio
        txt_files = [f for f in os.listdir(self.raw_texts_dir) if f.endswith('.txt')]

        if not txt_files:
            print("‚ùå No se encontraron archivos .txt en el directorio")
            return False

        txt_files.sort()  # Ordenar alfab√©ticamente

        print(f"üìÅ Encontrados {len(txt_files)} archivos .txt")

        # Diccionario de informaci√≥n de canciones de Soda Stereo
        songs_info = {
            "cuando_pase_el_temblor": {"titulo": "Cuando pase el temblor", "album": "Nada Personal", "year": "1985"},
            "nada_personal": {"titulo": "Nada personal", "album": "Nada Personal", "year": "1985"},
            "juegos_de_seduccion": {"titulo": "Juegos de seducci√≥n", "album": "Nada Personal", "year": "1985"},
            "imagenes_retro": {"titulo": "Im√°genes retro", "album": "Nada Personal", "year": "1985"},
            "danza_rota": {"titulo": "Danza rota", "album": "Nada Personal", "year": "1985"},
            "estoy_azulado": {"titulo": "Estoy azulado", "album": "Nada Personal", "year": "1985"},
            "profugos": {"titulo": "Pr√≥fugos", "album": "Signos", "year": "1986"},
            "persiana_americana": {"titulo": "Persiana americana", "album": "Signos", "year": "1986"},
            "signos": {"titulo": "Signos", "album": "Signos", "year": "1986"},
            "el_rito": {"titulo": "El rito", "album": "Signos", "year": "1986"},
            "no_existes": {"titulo": "No existes", "album": "Signos", "year": "1986"},
            "un_misil_en_mi_placard": {"titulo": "Un misil en mi placard", "album": "Soda Stereo", "year": "1984"},
            "te_hacen_falta_vitaminas": {"titulo": "Te hacen falta vitaminas", "album": "Soda Stereo", "year": "1984"},
            "mi_novia_tiene_biceps": {"titulo": "Mi novia tiene biceps", "album": "Soda Stereo", "year": "1984"},
            "tratame_suavemente": {"titulo": "Tr√°tame suavemente", "album": "Soda Stereo", "year": "1984"},
            "disco_eterno": {"titulo": "Disco Eterno", "album": "Soda Stereo", "year": "1984"},
            "en_la_ciudad_de_la_furia": {"titulo": "En la ciudad de la furia", "album": "Doble Vida", "year": "1988"},
            "lo_que_sangra": {"titulo": "Lo que sangra", "album": "Doble Vida", "year": "1988"},
            "corazon_delator": {"titulo": "Coraz√≥n delator", "album": "Doble Vida", "year": "1988"},
            "sobredosis_de_tv": {"titulo": "Sobredosis de TV", "album": "Doble Vida", "year": "1988"},
            "el_septimo_dia": {"titulo": "El s√©ptimo d√≠a", "album": "Canci√≥n Animal", "year": "1990"},
            "de_musica_ligera": {"titulo": "De m√∫sica ligera", "album": "Canci√≥n Animal", "year": "1990"},
            "te_para_tres": {"titulo": "T√© para tres", "album": "Canci√≥n Animal", "year": "1990"},
            "un_millon_de_anos_luz": {"titulo": "Un mill√≥n de a√±os luz", "album": "Canci√≥n Animal", "year": "1990"},
            "hombre_al_agua": {"titulo": "Hombre al agua", "album": "Canci√≥n Animal", "year": "1990"},
            "sueles_dejarme_solo": {"titulo": "Sueles dejarme solo", "album": "Canci√≥n Animal", "year": "1990"},
            "entre_canibales": {"titulo": "Entre can√≠bales", "album": "Canci√≥n Animal", "year": "1990"},
            "fue": {"titulo": "Fue", "album": "Dynamo", "year": "1992"},
            "paseando_por_roma": {"titulo": "Paseando por Roma", "album": "Sue√±o Stereo", "year": "1995"},
            "zoom": {"titulo": "Zoom", "album": "Sue√±o Stereo", "year": "1995"},
            "ella_uso_mi_cabeza_como_un_revolver": {"titulo": "Ella us√≥ mi cabeza como un revolver", "album": "Sue√±o Stereo", "year": "1995"}
        }

        # Crear lista de metadatos
        metadata_list = []

        for i, filename in enumerate(txt_files, 1):
            filepath = os.path.join(self.raw_texts_dir, filename)

            # Leer el contenido del archivo
            try:
                with open(filepath, 'r', encoding='utf-8') as f:
                    content = f.read().strip()
            except Exception as e:
                print(f"‚ö†Ô∏è  Error leyendo {filename}: {e}")
                continue

            # Extraer informaci√≥n del nombre del archivo
            clean_name = filename.replace('.txt', '')
            # Remover numeraci√≥n si existe (01_, 02_, etc.)
            clean_name = re.sub(r'^\d+_', '', clean_name)

            # Buscar informaci√≥n de la canci√≥n
            song_info = songs_info.get(clean_name, {})

            # Crear entrada de metadata
            metadata_entry = {
                'archivo': filename,
                'titulo': song_info.get('titulo', self.format_title_from_filename(clean_name)),
                'autor_fuente': 'Soda Stereo',
                'fecha': f"{song_info.get('year', '1985')}-01-01",
                'categoria': 'rock',
                'album': song_info.get('album', 'Desconocido'),
                'palabras_aprox': self.count_words(content)
            }

            metadata_list.append(metadata_entry)
            print(f"  {i:2d}. {metadata_entry['titulo']} - {metadata_entry['palabras_aprox']} palabras")

        # Guardar metadata
        df_metadata = pd.DataFrame(metadata_list)
        df_metadata.to_csv(self.metadata_file, index=False, encoding='utf-8')

        print(f"\n‚úÖ Metadata creado: {self.metadata_file}")
        print(f"üìä Total canciones procesadas: {len(metadata_list)}")

        return True

    def format_title_from_filename(self, clean_name):
        """Convierte nombre de archivo limpio a t√≠tulo legible"""
        title = clean_name.replace('_', ' ')
        # Capitalizar cada palabra
        words = []
        for word in title.split():
            if word.lower() in ['de', 'la', 'el', 'en', 'y', 'a', 'un', 'una', 'con', 'por', 'para', 'al', 'del']:
                words.append(word.lower())
            else:
                words.append(word.capitalize())

        # La primera palabra siempre va capitalizada
        if words:
            words[0] = words[0].capitalize()

        return ' '.join(words)

    def load_corpus(self):
        """Carga el corpus completo para an√°lisis"""
        if not os.path.exists(self.metadata_file):
            print("‚ùå No se encontr√≥ archivo de metadatos. Ejecuta scan_existing_corpus() primero.")
            return None, None

        metadata_df = pd.read_csv(self.metadata_file, encoding='utf-8')
        texts = {}

        for _, row in metadata_df.iterrows():
            filepath = os.path.join(self.raw_texts_dir, row['archivo'])
            if os.path.exists(filepath):
                try:
                    with open(filepath, 'r', encoding='utf-8') as f:
                        texts[row['archivo']] = f.read()
                except Exception as e:
                    print(f"‚ö†Ô∏è  Error leyendo {row['archivo']}: {e}")

        print(f"üìö Corpus cargado: {len(texts)} documentos")
        return texts, metadata_df

    def verify_corpus_integrity(self):
        """Verifica la integridad del corpus"""
        print("\nüîç Verificando integridad del corpus...")

        if not os.path.exists(self.metadata_file):
            print("‚ùå Archivo de metadatos no encontrado")
            return False

        metadata_df = pd.read_csv(self.metadata_file, encoding='utf-8')
        issues = []
        total_files = 0
        total_words = 0

        for _, row in metadata_df.iterrows():
            filepath = os.path.join(self.raw_texts_dir, row['archivo'])
            total_files += 1

            if not os.path.exists(filepath):
                issues.append(f"‚ùå Archivo faltante: {row['archivo']}")
                continue

            try:
                with open(filepath, 'r', encoding='utf-8') as f:
                    content = f.read().strip()

                if len(content) < 50:
                    issues.append(f"‚ö†Ô∏è  Archivo muy corto: {row['archivo']} ({len(content)} caracteres)")

                actual_words = self.count_words(content)
                total_words += actual_words

                # Verificar si el conteo de palabras es muy diferente
                if abs(actual_words - row['palabras_aprox']) > 50:
                    issues.append(f"‚ö†Ô∏è  Discrepancia en conteo de palabras: {row['archivo']} (metadata: {row['palabras_aprox']}, real: {actual_words})")

            except Exception as e:
                issues.append(f"‚ùå Error leyendo {row['archivo']}: {e}")

        if issues:
            print("üö® Problemas encontrados:")
            for issue in issues:
                print(f"  {issue}")
        else:
            print("‚úÖ Corpus √≠ntegro - no se encontraron problemas")

        print(f"üìä Total archivos verificados: {total_files}")
        print(f"üìù Total palabras en corpus: {total_words:,}")

        return len(issues) == 0

    def show_corpus_summary(self):
        """Muestra un resumen detallado del corpus creado"""
        if not os.path.exists(self.metadata_file):
            print("‚ùå No se encontr√≥ corpus para mostrar")
            return

        metadata_df = pd.read_csv(self.metadata_file, encoding='utf-8')

        print("\nüìã RESUMEN COMPLETO DEL CORPUS")
        print("=" * 60)

        # Resumen por √°lbum
        if 'album' in metadata_df.columns:
            album_counts = metadata_df.groupby('album').size().sort_values(ascending=False)

            print("üìÄ CANCIONES POR √ÅLBUM:")
            for album, count in album_counts.items():
                album_words = metadata_df[metadata_df['album'] == album]['palabras_aprox'].sum()
                print(f"  ‚Ä¢ {album}: {count} canciones ({album_words:,} palabras)")

        print(f"\nüìä ESTAD√çSTICAS GENERALES:")
        print(f"  ‚Ä¢ Total √°lbumes: {len(album_counts) if 'album' in metadata_df.columns else 0}")
        print(f"  ‚Ä¢ Total canciones: {len(metadata_df)}")
        print(f"  ‚Ä¢ Palabras totales: {metadata_df['palabras_aprox'].sum():,}")
        print(f"  ‚Ä¢ Promedio palabras/canci√≥n: {metadata_df['palabras_aprox'].mean():.0f}")
        print(f"  ‚Ä¢ Canci√≥n m√°s larga: {metadata_df.loc[metadata_df['palabras_aprox'].idxmax(), 'titulo']} ({metadata_df['palabras_aprox'].max()} palabras)")
        print(f"  ‚Ä¢ Canci√≥n m√°s corta: {metadata_df.loc[metadata_df['palabras_aprox'].idxmin(), 'titulo']} ({metadata_df['palabras_aprox'].min()} palabras)")

        # Lista completa numerada
        print(f"\nüìù LISTA COMPLETA DE CANCIONES:")
        for i, (_, row) in enumerate(metadata_df.iterrows(), 1):
            album_short = row.get('album', 'N/A')[:20] + "..." if len(str(row.get('album', 'N/A'))) > 20 else row.get('album', 'N/A')
            print(f"  {i:2d}. {row['titulo']} - {album_short} ({row['palabras_aprox']} palabras)")

    def mostrar_fragmentos_por_cancion(self, palabras_por_fragmento=15):
        """Muestra un fragmento representativo de cada canci√≥n"""
        if not os.path.exists(self.metadata_file):
            print("‚ùå Metadata no encontrado")
            return False

        metadata_df = pd.read_csv(self.metadata_file, encoding='utf-8')
        texts, _ = self.load_corpus()

        if not texts:
            print("‚ùå No se pudieron cargar los textos")
            return False

        print("\nüéµ FRAGMENTO REPRESENTATIVO POR CANCI√ìN")
        print("=" * 80)

        # Ordenar todas las canciones por √°lbum y t√≠tulo
        metadata_df_sorted = metadata_df.sort_values(['album', 'titulo'])

        contador_cancion = 0
        album_actual = None

        for _, song in metadata_df_sorted.iterrows():
            # Mostrar encabezado de √°lbum cuando cambie
            if album_actual != song['album']:
                album_actual = song['album']
                print(f"\nüìÄ {album_actual.upper()}")
                print("-" * 60)

            if song['archivo'] in texts:
                contador_cancion += 1
                texto = texts[song['archivo']]

                # Dividir en l√≠neas v√°lidas
                lineas = [s.strip() for s in texto.split('\n') if len(s.strip()) > 10]

                if lineas:
                    # Tomar primera l√≠nea significativa
                    linea_elegida = lineas[0]
                    for linea in lineas:
                        if len(linea.split()) >= 5:
                            linea_elegida = linea
                            break

                    palabras = linea_elegida.split()[:palabras_por_fragmento]
                    fragmento_corto = ' '.join(palabras)

                    print(f"  {contador_cancion:2d}. '{song['titulo']}':")
                    print(f"      \"{fragmento_corto}{'...' if len(palabras) == palabras_por_fragmento else ''}\"")
                    print()

        print(f"üìä Total de canciones mostradas: {contador_cancion}")
        return True

    def crear_nube_palabras_por_album(self):
        """Crea nubes de palabras para cada √°lbum"""
        try:
            if not os.path.exists(self.metadata_file):
                print("‚ùå Metadata no encontrado")
                return False

            metadata_df = pd.read_csv(self.metadata_file, encoding='utf-8')
            texts, _ = self.load_corpus()

            if not texts:
                print("‚ùå No se pudieron cargar los textos")
                return False

            print("\n‚òÅÔ∏è CREANDO NUBES DE PALABRAS POR √ÅLBUM...")

            # Stop words en espa√±ol
            stop_words_es = {
                'que', 'de', 'la', 'el', 'en', 'y', 'a', 'un', 'es', 'se', 'no', 'te',
                'lo', 'le', 'da', 'su', 'por', 'son', 'con', 'para', 'las', 'del',
                'una', 'me', 'al', 'como', 'mi', 'tu', 'si', 'los', 'pero', 'sus',
                'ya', 'ni', 'muy', 'tan', 'm√°s', 'vez', 'ser', 'sin', 'otro', 'otra',
                'cuando', 'donde', 'quien', 'cual', 'todo', 'todos', 'cada', 'hasta',
                'desde', 'entre', 'sobre', 'bajo', 'tras', 'durante', 'mediante', 'esta', 'estos', 'este', 'cuando', 'estas', 'con', 'cabe',
            }

            # Agrupar textos por √°lbum
            albums = metadata_df['album'].unique()
            num_albums = len(albums)

            # Crear figura con subplots
            cols = 2 if num_albums > 2 else num_albums
            rows = (num_albums + cols - 1) // cols

            fig, axes = plt.subplots(rows, cols, figsize=(15, 5*rows))
            if num_albums == 1:
                axes = [axes]
            elif rows == 1:
                axes = axes if num_albums > 1 else [axes]
            else:
                axes = axes.flatten()

            colores_albums = ['viridis', 'plasma', 'inferno', 'magma', 'cividis', 'spring', 'summer']

            for i, album in enumerate(albums):
                # Combinar todos los textos del √°lbum
                album_songs = metadata_df[metadata_df['album'] == album]
                texto_album = ""

                for _, song in album_songs.iterrows():
                    if song['archivo'] in texts:
                        texto_album += " " + texts[song['archivo']]

                if texto_album.strip():
                    # Crear wordcloud
                    wordcloud = WordCloud(
                        width=800,
                        height=400,
                        background_color='white',
                        stopwords=stop_words_es,
                        colormap=colores_albums[i % len(colores_albums)],
                        max_words=100,
                        relative_scaling=0.5,
                        collocations=False
                    ).generate(texto_album.lower())

                    # Mostrar en subplot
                    ax = axes[i] if num_albums > 1 else axes[0]
                    ax.imshow(wordcloud, interpolation='bilinear')
                    ax.set_title(f'‚òÅÔ∏è {album}', fontsize=14, fontweight='bold', pad=20)
                    ax.axis('off')

                    print(f"‚úÖ Nube creada para: {album}")

            # Ocultar subplots vac√≠os
            for j in range(i + 1, len(axes)):
                axes[j].axis('off')

            plt.tight_layout()
            plt.suptitle('NUBES DE PALABRAS POR √ÅLBUM - SODA STEREO',
                        fontsize=16, fontweight='bold', y=0.98)

            # Guardar
            viz_dir = os.path.join(self.base_dir, 'visualizations')
            os.makedirs(viz_dir, exist_ok=True)
            wordcloud_path = os.path.join(viz_dir, 'wordclouds_por_album.png')
            plt.savefig(wordcloud_path, dpi=300, bbox_inches='tight', facecolor='white')
            plt.show()

            print(f"üíæ Nubes de palabras guardadas en: {wordcloud_path}")
            return True

        except ImportError:
            print("‚ùå Error: Instala wordcloud con: pip install wordcloud")
            return False
        except Exception as e:
            print(f"‚ùå Error creando nubes de palabras: {e}")
            return False

    def export_metadata_format(self):
        """Exporta el metadata en formato est√°ndar limpio"""
        if not os.path.exists(self.metadata_file):
            print("‚ùå No se encontr√≥ archivo de metadatos")
            return False

        # Cargar metadata existente
        metadata_df = pd.read_csv(self.metadata_file, encoding='utf-8')

        # Crear versi√≥n con solo columnas obligatorias y formato est√°ndar
        standard_metadata = metadata_df[['archivo', 'titulo', 'autor_fuente', 'fecha', 'categoria', 'palabras_aprox']].copy()

        # Exportar versi√≥n est√°ndar
        standard_file = os.path.join(self.base_dir, "metadata_standard.csv")
        standard_metadata.to_csv(standard_file, index=False, encoding='utf-8')

        print(f"‚úÖ Metadata est√°ndar exportado: {standard_file}")
        print("\nüìã FORMATO EST√ÅNDAR CREADO:")
        print("Columnas: archivo, titulo, autor_fuente, fecha, categoria, palabras_aprox")
        print(f"Registros: {len(standard_metadata)}")

        return True

# Funciones de an√°lisis mejoradas
def analyze_corpus_stats(texts, metadata_df):
    """Genera estad√≠sticas detalladas del corpus"""
    print("\nüìà AN√ÅLISIS ESTAD√çSTICO DETALLADO")
    print("=" * 60)

    if not texts:
        print("‚ùå No hay textos para analizar")
        return

    total_words = sum(len(re.findall(r'\b\w+\b', text.lower())) for text in texts.values())
    total_chars = sum(len(text) for text in texts.values())
    total_docs = len(texts)

    print(f"üìÑ Total documentos: {total_docs}")
    print(f"üìù Total palabras: {total_words:,}")
    print(f"üî§ Total caracteres: {total_chars:,}")
    print(f"üìä Promedio palabras/canci√≥n: {total_words // total_docs if total_docs > 0 else 0}")
    print(f"üìè Promedio caracteres/canci√≥n: {total_chars // total_docs if total_docs > 0 else 0}")

    # An√°lisis temporal si hay informaci√≥n de √°lbumes
    if 'album' in metadata_df.columns and 'fecha' in metadata_df.columns:
        print(f"\nüìÖ EVOLUCI√ìN TEMPORAL:")
        metadata_df['year'] = pd.to_datetime(metadata_df['fecha']).dt.year
        yearly_stats = metadata_df.groupby('year').agg({
            'titulo': 'count',
            'palabras_aprox': ['sum', 'mean']
        }).round(0)

        for year in sorted(metadata_df['year'].unique()):
            year_data = metadata_df[metadata_df['year'] == year]
            print(f"  ‚Ä¢ {year}: {len(year_data)} canciones, {year_data['palabras_aprox'].sum():,} palabras")

    # Top 20 palabras m√°s comunes (excluyendo stop words)
    print(f"\nüî§ TOP 20 PALABRAS M√ÅS FRECUENTES:")
    all_text = ' '.join(texts.values()).lower()
    words = re.findall(r'\b\w+\b', all_text)

    # Stop words extendida
    stop_words = {
         'que', 'a', 'ante', 'bajo', 'esta', 'estos', 'este', 'cuando', 'estas', 'con', 'cabe', 'de', 'la', 'el', 'en', 'y', 'a', 'un', 'es', 'se', 'no', 'te',
                'lo', 'le', 'da', 'su', 'por', 'son', 'con', 'para', 'las', 'del', 'fue',
                'una', 'me', 'al', 'como', 'mi', 'tu', 'si', 'los', 'pero', 'sus',
                'ya', 'ni', 'muy', 'tan', 'm√°s', 'vez', 'ser', 'sin', 'otro', 'otra',
                'cuando', 'donde', 'quien', 'cual', 'todo', 'todos', 'cada', 'hasta',
                'desde', 'entre', 'sobre', 'bajo', 'tras', 'durante', 'mediante'
    }

    filtered_words = [word for word in words if len(word) > 2 and word not in stop_words]
    word_freq = Counter(filtered_words)

    for i, (word, count) in enumerate(word_freq.most_common(20), 1):
        percentage = (count / len(filtered_words)) * 100
        print(f"  {i:2d}. '{word}': {count} veces ({percentage:.1f}%)")

def create_corpus_visualization(texts, metadata_df):
    """Crea visualizaci√≥n completa del corpus"""
    try:
        import matplotlib.pyplot as plt
        import seaborn as sns

        # Configurar estilo
        plt.style.use('default')
        fig = plt.figure(figsize=(16, 12))

        # Crear grid de subplots
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

        # 1. Canciones por √°lbum (si existe info de √°lbum)
        ax1 = fig.add_subplot(gs[0, 0])
        if 'album' in metadata_df.columns:
            album_counts = metadata_df['album'].value_counts()
            colors = plt.cm.Set3(range(len(album_counts)))
            ax1.pie(album_counts.values, labels=[album[:15] + '...' if len(album) > 15 else album for album in album_counts.index],
                   autopct='%1.0f', colors=colors, textprops={'fontsize': 8})
            ax1.set_title('Distribuci√≥n por √Ålbum', fontsize=12, fontweight='bold')

        # 2. Palabras por √°lbum
        ax2 = fig.add_subplot(gs[0, 1])
        if 'album' in metadata_df.columns:
            album_words = metadata_df.groupby('album')['palabras_aprox'].sum().sort_values(ascending=True)
            ax2.barh(range(len(album_words)), album_words.values, color='lightcoral')
            ax2.set_title('Palabras Totales por √Ålbum', fontsize=12, fontweight='bold')
            ax2.set_xlabel('Palabras')
            ax2.set_yticks(range(len(album_words)))
            ax2.set_yticklabels([album[:20] for album in album_words.index])

        # 3. Evoluci√≥n temporal
        ax3 = fig.add_subplot(gs[0, 2])
        if 'fecha' in metadata_df.columns:
            metadata_df['year'] = pd.to_datetime(metadata_df['fecha']).dt.year
            yearly_avg = metadata_df.groupby('year')['palabras_aprox'].mean()
            yearly_count = metadata_df.groupby('year').size()

            ax3_twin = ax3.twinx()
            line1 = ax3.plot(yearly_avg.index, yearly_avg.values, 'bo-', label='Promedio palabras', linewidth=2)
            bar1 = ax3_twin.bar(yearly_count.index, yearly_count.values, alpha=0.3, color='red', label='Num. canciones')

            ax3.set_title('Evoluci√≥n Temporal', fontsize=12, fontweight='bold')
            ax3.set_xlabel('A√±o')
            ax3.set_ylabel('Promedio de palabras', color='blue')
            ax3_twin.set_ylabel('N√∫mero de canciones', color='red')
            ax3.grid(True, alpha=0.3)

        # 4. Distribuci√≥n de longitud
        ax4 = fig.add_subplot(gs[1, 0])
        word_counts = [len(re.findall(r'\b\w+\b', text.lower())) for text in texts.values()]
        ax4.hist(word_counts, bins=12, alpha=0.7, color='lightgreen', edgecolor='black')
        ax4.set_title('Distribuci√≥n de Longitud', fontsize=12, fontweight='bold')
        ax4.set_xlabel('Palabras por canci√≥n')
        ax4.set_ylabel('Frecuencia')
        ax4.axvline(np.mean(word_counts), color='red', linestyle='--', label=f'Promedio: {np.mean(word_counts):.0f}')
        ax4.legend()

        # 5. Top palabras m√°s frecuentes
        ax5 = fig.add_subplot(gs[1, 1:])
        all_text = ' '.join(texts.values()).lower()
        words = re.findall(r'\b\w+\b', all_text)
        stop_words = {
            'que', 'este', 'donde', 'sin', 'cuando', 'todos', 'est√°', 'mucho', 'muchas', 'siempre', 'todo', 'esta', 'de', 'la', 'el', 'en', 'y', 'a', 'un', 'es', 'se', 'no', 'te',
            'lo', 'le', 'da', 'su', 'por', 'son', 'con', 'para', 'las', 'del',
            'una', 'me', 'al', 'como', 'mi', 'tu', 'si', 'los', 'pero', 'sus'
        }
        filtered_words = [word for word in words if len(word) > 2 and word not in stop_words]
        word_freq = Counter(filtered_words).most_common(15)

        words_top, counts_top = zip(*word_freq)
        ax5.barh(range(len(words_top)), counts_top, color='skyblue')
        ax5.set_title('Top 15 Palabras M√°s Frecuentes', fontsize=12, fontweight='bold')
        ax5.set_xlabel('Frecuencia')
        ax5.set_yticks(range(len(words_top)))
        ax5.set_yticklabels(words_top)

        # 6. Estad√≠sticas generales (texto)
        ax6 = fig.add_subplot(gs[2, :])
        ax6.axis('off')

        # Calcular estad√≠sticas
        total_words = sum(len(re.findall(r'\b\w+\b', text.lower())) for text in texts.values())
        total_chars = sum(len(text) for text in texts.values())
        avg_words = total_words / len(texts) if texts else 0

        stats_text = f"""
üìä ESTAD√çSTICAS GENERALES DEL CORPUS
‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê
üìÅ Total de documentos: {len(texts)}
üìù Total de palabras: {total_words:,}
üî§ Total de caracteres: {total_chars:,}
üìä Promedio palabras/canci√≥n: {avg_words:.0f}
üìè Promedio caracteres/canci√≥n: {total_chars // len(texts) if texts else 0:,}
üìÄ √Ålbumes √∫nicos: {metadata_df['album'].nunique() if 'album' in metadata_df.columns else 'N/A'}
üéµ Per√≠odo: {metadata_df['fecha'].min()[:4] if 'fecha' in metadata_df.columns else 'N/A'} - {metadata_df['fecha'].max()[:4] if 'fecha' in metadata_df.columns else 'N/A'}
        """

        ax6.text(0.1, 0.5, stats_text, fontsize=11, verticalalignment='center',
                fontfamily='monospace', bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgray", alpha=0.8))

        plt.suptitle('AN√ÅLISIS COMPLETO DEL CORPUS - SODA STEREO', fontsize=16, fontweight='bold', y=0.98)

        # Guardar visualizaci√≥n
        corpus_base = "/content/drive/MyDrive/Programaci√≥n de lenguaje/Trabajo Integrador PNL/mi_corpus"
        viz_dir = os.path.join(corpus_base, 'visualizations')
        os.makedirs(viz_dir, exist_ok=True)
        viz_path = os.path.join(viz_dir, 'corpus_analysis_complete.png')
        plt.savefig(viz_path, dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()

        print(f"üìä An√°lisis visual completo guardado en: {viz_path}")

    except ImportError:
        print("üìä Para visualizaciones completas, instala: pip install matplotlib seaborn")
    except Exception as e:
        print(f"‚ö†Ô∏è  Error en visualizaci√≥n: {e}")

# Funci√≥n principal optimizada
def main():
    """Se utiliza el metadata existente"""
    print("üéµ CORPUS ANALYZER - SODA STEREO")
    print("=" * 60)

    correct_path = "/content/drive/MyDrive/Programaci√≥n de lenguaje/Trabajo Integrador PNL/mi_corpus"
    print(f"üéØ Analizando corpus existente en: {correct_path}/raw_texts")
    print("=" * 60)

    builder = CorpusBuilder(base_dir=correct_path)

    # Verifica si METADATA EXISTE
    if os.path.exists(builder.metadata_file):
        print("‚úÖ Metadata existente ha sido encontrado")

        # Carga el doc y lo analiza
        texts, metadata_df = builder.load_corpus()
        if texts:
            builder.verify_corpus_integrity()
            builder.show_corpus_summary()
            analyze_corpus_stats(texts, metadata_df)

            # NUEVAS FUNCIONALIDADES - agregar aqu√≠:
            print("\nüéµ Mostrando fragmentos representativos...")
            builder.mostrar_fragmentos_por_cancion(palabras_por_fragmento=20)

            print("\n‚òÅÔ∏è Creando nubes de palabras por √°lbum...")
            builder.crear_nube_palabras_por_album()

            create_corpus_visualization(texts, metadata_df)

            print(f"\nüéâ ¬°AN√ÅLISIS COMPLETADO!")
    else:
        print("‚ùå Metadata no ha sido encontrado! Se generar√° metadata")
        # Escanear y crear metadata
        if builder.scan_existing_corpus():
            texts, metadata_df = builder.load_corpus()
            if texts:
                builder.verify_corpus_integrity()
                builder.show_corpus_summary()
                analyze_corpus_stats(texts, metadata_df)

                # NUEVAS FUNCIONALIDADES:
                print("\nüéµ Mostrando fragmentos representativos...")
                builder.mostrar_fragmentos_por_cancion(palabras_por_fragmento=20)

                print("\n‚òÅÔ∏è Creando nubes de palabras por √°lbum...")
                builder.crear_nube_palabras_por_album()

                create_corpus_visualization(texts, metadata_df)

                print(f"\nüéâ ¬°AN√ÅLISIS COMPLETADO!")

    return builder

if __name__ == "__main__":
    main()
```
